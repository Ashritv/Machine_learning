```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

//x=np.random.randint(0,10,5).reshape(-1,1)
y=np.random.randint(10,20,5)//  random number generation

```python
z=pd.read_excel('excelfile.xlsx')
s=pd.read_excel('excelfile.xlsx')
df=pd.DataFrame(z)
a=df.loc[:,"x"]
b=df.loc[:,"y"]
a=np.array(a).reshape(-1,1)
```

```python
plt.scatter(a,b)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
slr=LinearRegression()
```

```python
slr.fit(a,b)
```

```python
pred=slr.predict(a)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(b,pred)
plt.plot(a,pred,color='red',marker='*')
print("mse",mse)
print("intercept",slr.intercept_)
print("slope",slr.coef_)
```
